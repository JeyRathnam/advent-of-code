```python
import os 
def passportValidation(fileName):
    with open(os.path.join(os.path.abspath(''), fileName), 'r') as file:
        defaultPassport = {
            "byr" : "",
            "iyr" : "",
            "eyr" : "",
            "hgt" : "",
            "hcl" : "",
            "ecl" : "",
            "pid" : "",
            "cid" : "DEFAULT"
        }
        validCount = 0
        line = 0
        lines = file.read().splitlines()

        joinedLines = []
        curr = defaultPassport.copy()

        for i, line in enumerate(lines):
            splitLine = line.split(" ")
            for kv in splitLine:
                values = kv.split(":")
                if len(values) == 2 and values[0] in defaultPassport:
                    curr[values[0]] = values[1]
            
            if line == "" or i == len(lines)-1:
                if all(len(val) > 0 for val in curr.values()):
                    validCount += 1
                curr = defaultPassport.copy()
                
        return validCount
                
```

```python
import re
import os 
def validateRange(input, lb, up):
    return input >= lb and input <= up

def validateByr(input):
    return validateRange(int(input), 1920, 2002)

def validateIyr(input):
    return validateRange(int(input), 2010, 2020)

def validateEyr(input):
    return validateRange(int(input), 2020, 2030)

def validateHgt(input):
    heightType = input[-2:]
    height = input[:-2]
    if not height or not heightType:
        return False
    height = int(height)
    if heightType == "cm":
        return height >= 150 and height <= 193
    elif heightType == "in":
        return height >= 59 and height <= 76
    else:
        return False
    
def validateHcl(input):
    pattern = re.compile("^#[0-9a-z]{6}$")
    return bool(pattern.match(input))

def validateEcl(input):
    eyeColors = {
        'amb' : "amb",
        'blu' : 'blu',
        'brn' : 'brn',
        'gry' : 'gry',
        'grn' : 'grn',
        'hzl' : 'hzl',
        'oth' : 'oth'
    }
    
    if input in eyeColors:
        return True
    else:
        return False
    
def validatePid(input):
    pattern = re.compile("^[0-9]{9}$")
    return bool(pattern.match(input))


def validateCid(input):
    return True

def passportValidation2(fileName):
    validator = {
        'byr' : validateByr,
        'iyr' : validateIyr,
        'eyr' : validateEyr,
        'hgt' : validateHgt,
        'hcl' : validateHcl,
        'ecl' : validateEcl,
        'pid' : validatePid,
        'cid' : validateCid
    }
    
    defaultPassport = {
            "byr" : "",
            "iyr" : "",
            "eyr" : "",
            "hgt" : "",
            "hcl" : "",
            "ecl" : "",
            "pid" : "",
            "cid" : True
        }
    with open(os.path.join(os.path.abspath(''), fileName), 'r') as file:
        validCount = 0
        lines = file.read().splitlines()
        i = 0
        isValid = True
        curr = defaultPassport.copy()
        while i < len(lines):
            splitLine = lines[i].split(" ")
            for kv in splitLine:
                keyValue = kv.split(":")
                if len(keyValue) == 2 and keyValue[0] in validator:
                    print(keyValue)
                    if not validator[keyValue[0]](keyValue[1]):
                        isValid = False
                        while i < len(lines) -1 and lines[i] != "":
                            i += 1
                    else:
                        curr[keyValue[0]] = True
            
            
            if lines[i]  == "" or i == len(lines)-1:
                print(curr)
                if isValid and all(val == True for val in curr.values()):
                    validCount += 1
                isValid = True
                curr = defaultPassport.copy()
            i += 1
        return validCount
```
