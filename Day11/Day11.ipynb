```python
def getNeighbouringOccupancy(x, y, lines, debug=False):
    occupancy = 0
    
    #left
    if y-1 >= 0 and lines[x][y-1] == '#':
        occupancy += 1
        if debug:
            print('Left', (x, y-1))
        
    #top
    if x-1 >= 0 and lines[x-1][y] == '#':
        occupancy += 1
        if debug:
            print('top')
        
    #right
    if y+1 <= len(lines[0])-1 and lines[x][y+1] == '#':
        occupancy += 1
        if debug:
            print('right')
        
    #bottom
    if x+1 <= len(lines)-1 and lines[x+1][y] == '#':
        occupancy += 1
        if debug:
            print('bottom')
        
    #top left
    if x-1 >= 0 and y-1 >= 0 and lines[x-1][y-1] == '#':
        occupancy += 1
        if debug:
            print('top left')
        
    #top right 
    if x-1 >=0 and y+1 <=len(lines[0])-1 and lines[x-1][y+1] == '#':
        occupancy += 1 
        if debug:
            print('top right')
        
    #bottom left
    if x+1 <= len(lines)-1 and y-1 >= 0 and lines[x+1][y-1] == '#':
        occupancy += 1
        if debug:
            print('bottom left')
        
    #bottom right
    if x+1 <= len(lines)-1 and y+1 <= len(lines[0])-1 and lines[x+1][y+1] == '#':
        occupancy += 1
        if debug:
            print('bottom right')
        
    return occupancy
def printSeats(seats):
    for x in range(len(seats[0])):
        for y in range(len(seats)):
            print(seats[x][y], end="")
        print()
def compareTwoLists(l1, l2):
    if len(l1) != len(l2):
        return False
    
    for x in range(len(l1)):
        for y in range(len(l1[x])):
            if l1[x][y] != l2[x][y]:
                return False
    return True

def countSeats(seats):
    count = 0
    for x in range(len(seats)):
        for y in range(len(seats[0])):
            if seats[x][y] == '#':
                count += 1
    return count
```

```python
import os
def seatingSeat(lines):
    output = []
    row = ['.'] * len(lines[0])
    for x in range(len(lines)):
            output.append(row[:])
    for x in range(len(lines)):
        for y in range(len(lines[0])):
            if lines[x][y] == 'L':
                if getNeighbouringOccupancy(x, y, lines) == 0:
                    output[x][y] = '#'
                else:
                    output[x][y] = 'L'

            elif lines[x][y] == '#':
                if getNeighbouringOccupancy(x, y, lines) >= 4:
                    output[x][y] = 'L'
                else:
                    output[x][y] = '#'
                    
    if compareTwoLists(output, lines):
        print(countSeats(output))
        return output
    else:
        seatingSeat(output)

filename = 'input.txt'
with open(os.path.join(os.path.abspath(''), filename), 'r') as f:
        lines = f.read().splitlines()
        print(seatingSeat(lines))
#         print(getNeighbouringOccupancy(0, 4, lines))
        
```
