```python
def getNeighbouringOccupancy(x, y, lines, debug=False):
    occupancy = 0
    
    #left
    ty = y
    while ty-1 >= 0:
        if lines[x][ty-1] == '#':
            occupancy += 1
            if debug:
                print('Left', (x, ty-1))
            break
        elif lines[x][ty-1] == 'L':
            break
            
        ty -= 1
        
    #top
    tx = x
    while tx-1 >= 0:
        if lines[tx-1][y] == '#':
            occupancy += 1
            if debug:
                print('top', (tx-1, y))
            break
        elif lines[tx-1][y] == 'L':
            break
        tx -= 1
        
    #right
    ty = y
    while ty+1 <= len(lines[0])-1:
        if lines[x][ty+1] == '#':
            occupancy += 1
            if debug:
                print('right', (x, ty+1))
            break
        elif lines[x][ty+1] == 'L':
            break
        ty += 1
        
    #bottom
    tx = x
    while tx+1 <= len(lines)-1:
        if lines[tx+1][y] == '#':
            occupancy += 1
            if debug:
                print('bottom', (tx+1, y))
            break
        elif lines[tx+1][y] == 'L':
            break
        tx += 1
        
        
        
    #top left
    tx = x
    ty = y
    while tx-1 >= 0 and ty-1 >= 0:
        if lines[tx-1][ty-1] == '#':
            occupancy += 1
            if debug:
                print('top left', (tx-1, ty-1))
            break
        elif lines[tx-1][ty-1] == 'L':
            break
        tx -= 1
        ty -= 1
        
        
    #top right 
    tx = x
    ty = y
    while tx-1 >=0 and ty+1 <=len(lines[0])-1:
        if lines[tx-1][ty+1] == '#':
            occupancy += 1 
            if debug:
                print('top right', (tx-1, ty+1))
            break
        elif lines[tx-1][ty+1] == 'L':
            break
        tx -= 1
        ty += 1
        
    #bottom left
    tx = x
    ty = y
    while tx+1 <= len(lines)-1 and ty-1 >= 0:
        if lines[tx+1][ty-1] == '#':
            occupancy += 1
            if debug:
                print('bottom left', (tx+1, ty-1))
            break
        elif lines[tx+1][ty-1] == 'L':
            break
        tx += 1
        ty -= 1
        
    #bottom right
    tx = x
    ty = y
    while tx+1 <= len(lines)-1 and ty+1 <= len(lines[0])-1:
        if lines[tx+1][ty+1] == '#':
            occupancy += 1
            if debug:
                print('bottom right', (tx+1, ty+1))
            break
        elif lines[tx+1][ty+1] == 'L':
            break
        tx += 1
        ty += 1
        
    return occupancy

def printSeats(seats):
    for x in range(len(seats[0])):
        for y in range(len(seats)):
            print(seats[x][y], end="")
        print()
def compareTwoLists(l1, l2):
    if len(l1) != len(l2):
        return False
    
    for x in range(len(l1)):
        for y in range(len(l1[x])):
            if l1[x][y] != l2[x][y]:
                return False
    return True

def countSeats(seats):
    count = 0
    for x in range(len(seats)):
        for y in range(len(seats[0])):
            if seats[x][y] == '#':
                count += 1
    return count



def topRight(x, y, lines, debug=False):
    tx = x
    ty = y
    occupancy = 0
    while tx-1 >=0 and ty+1 <=len(lines[0])-1:
        if lines[tx-1][ty+1] == '#':
            occupancy += 1 
            if debug:
                print('top right', (tx-1, ty+1))
            break
        elif lines[tx-1][ty+1] == 'L':
            break
        tx -= 1
        ty += 1
    return occupancy
```
