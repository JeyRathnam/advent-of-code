```python
import os

NESW = "NESW"

class RainRisk:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.wx = 0
        self.wy = 0
        self.dir = "E"
        
    def N(self, value):
        self.wy += value
        
    def S(self, value):
        self.wy -= value
        
    def W(self, value):
        self.wx -= value
        
    def E(self, value):
        self.wx += value
        
    def F(self, value):
#         self.call(self.dir, value)
        self.x = (self.wx * value) + self.x
        self.y = (self.wy * value)  + self.y
        
    def R(self, value):        
        if value == 90:
            x = self.wx
            y = self.wy
            self.wy = x * -1
            self.wx = y
        elif value == 180:
            x = self.wx
            y = self.wy
            self.wx = x * -1
            self.wy = y * -1
        elif value == 270:
            x = self.wx
            y = self.wy
            self.wx = y * -1
            self.wy = x
        
    def L(self, value):
        if value == 90:
            x = self.wx
            y = self.wy
            self.wx = y * -1
            self.wy = x
        elif value == 180:
            x = self.wx
            y = self.wy
            self.wx = x * -1
            self.wy = y * -1
        elif value == 270:
            x = self.wx
            y = self.wy
            self.wx = y
            self.wy = x * -1
        
    def calls(self):
        return {
            'N' : self.N,
            'S' : self.S,
            'W' : self.W,
            'E' : self.E,
            'F' : self.F,
            'R' : self.R,
            'L' : self.L
        }
    
    def call(self, name, value):
        if name not in self.calls():
            return
        
        self.calls()[name](value)
        
    def sum(self):
        return abs(self.x) + abs(self.y)
        
        
obj = RainRisk()
obj.wx = 10
obj.wy = 1
fileName = "input.txt" 
with open(os.path.join(os.path.abspath(''), fileName), 'r') as file:     
    lines = file.read().splitlines()
    for line in lines:
        d = line[0]
        value = int(line[1:])
#         print(d)
        obj.call(d, value)
        print(d + str(value), (obj.wx, obj.wy), (obj.x, obj.y),  obj.dir)
            
# print(obj.x, obj.y)
print(obj.sum())
```
