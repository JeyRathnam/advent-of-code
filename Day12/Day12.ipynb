```python
import os

def test(a):
    print("test" + a)
    
def test2():
    print("test 2")
a = {
    'F' : test,
    'F2' : test2
}

NESW = "NESW"

class RainRisk:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.dir = "E"
        
    def N(self, value):
        self.y += value
        
    def S(self, value):
        self.y -= value
        
    def W(self, value):
        self.x -= value
        
    def E(self, value):
        self.x += value
        
    def F(self, value):
        self.call(self.dir, value)
        
    def R(self, value):
        curr = NESW.find(self.dir)
        
        i = (value // 90) % 4
    
        while i > 0:
            curr += 1
            if curr >= len(NESW):
                curr = 0
            i -= 1
        self.dir = NESW[curr]
        
    def L(self, value):
        curr = NESW.find(self.dir)
        i = (value // 90) % 4
    
        while i > 0:
            curr -= 1
            if curr < 0:
                curr = len(NESW) - 1
            i -= 1
        self.dir = NESW[curr]
        
    def calls(self):
        return {
            'N' : self.N,
            'S' : self.S,
            'W' : self.W,
            'E' : self.E,
            'F' : self.F,
            'R' : self.R,
            'L' : self.L
        }
    
    def call(self, name, value):
        if name not in self.calls():
            return
        
        self.calls()[name](value)
        
    def sum(self):
        return abs(self.x) + abs(self.y)
        
        
obj = RainRisk()

fileName = "input.txt" 
with open(os.path.join(os.path.abspath(''), fileName), 'r') as file:     
    lines = file.read().splitlines()
    for line in lines:
        d = line[0]
        value = int(line[1:])
#         print(d)
        obj.call(d, value)
#         print(d + str(value), obj.x, obj.y, obj.dir)
            
# print(obj.x, obj.y)
print(obj.sum())
```
